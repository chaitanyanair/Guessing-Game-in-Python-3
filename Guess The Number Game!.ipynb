```python
import random 
num = random.randint(1,100)

print("Welcome to the GUESSING GAME!")
print("You will guess a number between 1 to 100")
print("If you guess the number generated by the computer correctly, you WIN!")
print("If your guess is within 10 digits of the correct number, we will tell you you are WARM")
print("If your furher guesses are closer that 10 digits, we will tell you you are WARMER")
print("If your consequtive guesses are farther than 10 digits, we will say you are COLDER")
print("If your guess is NOT within 10 digits of the correct number, we will tell you you are COLD")
print("You will enter guesses until you have guessed the number correctly")

guesses = [0]

while True:

    # we can copy the code from above to take an input
    guess = int(input("I'm thinking of a number between 1 and 100.\n  What is your guess? "))
    
    if guess < 1 or guess > 100:
        print('OUT OF BOUNDS! Please try again: ')
        continue
    
    # here we compare the player's guess to our number
    if guess == num:
        print(f'CONGRATULATIONS, YOU GUESSED IT IN ONLY {len(guesses)} GUESSES!!')
        break
        
    # if guess is incorrect, add guess to the list
    guesses.append(guess)
    
    # when testing the first guess, guesses[-2]==0, which evaluates to False
    # and brings us down to the second section
    
    if guesses[-2]:  
        if abs(num-guess) < abs(num-guesses[-2]):
            print('WARMER!')
        else:
            print('COLDER!')
   
    else:
        if abs(num-guess) <= 10:
            print('WARM!')
        else:
            print('COLD!')
```
